```python
df['engine_cylinders'] = df['engine_cylinders'].fillna(
    df['engine_cylinders'].mode()[0]
)
```

```python
df = df.drop(['number_of_doors', 'market_category','make','model','engine_fuel_type','vehicle_style'], axis=1)
```

```python
df['hourse'] = df['hourse'].fillna(
    df['hourse'].mode()[0]
)
```

```python
df = pd.get_dummies(
      df,
          columns=['transmission_type', 'driven_wheels', 'vehicle_size']
          )
```
